```java
import java.util.Scanner;
import java.util.HashMap;
import java.util.ArrayList;

public class Menu {
    Scanner userInput = new Scanner(System.in);
    Scanner userInput2 = new Scanner(System.in);
    HashMap<String, Long> userPhoneBook = new HashMap<String, Long>();

    public void phoneBook() {
        System.out.println("");
        System.out.println("Phone Book");
        System.out.println("----------");
        System.out.println("1) Add to phone book");
        System.out.println("2) Access phone contact");
        System.out.println("3) Remove phone contact");
        System.out.println("4) View your phone book");
        System.out.println("0) Return to main menu");
        int phoneBookSelection = userInput.nextInt();
        if (phoneBookSelection == 1) {
            System.out.println("Enter the name of the phone contact:");
            String contactName = userInput2.nextLine();
            System.out.println("Enter their phone number:");
            long phoneNumber = userInput.nextLong();
            userPhoneBook.put(contactName, phoneNumber);
            phoneBook();
        } else if (phoneBookSelection == 2) {
            System.out.println("Enter the name of the phone contact you wish to access:");
            String contactNameAccess = userInput2.nextLine();
            System.out.println("Output:");
            for (String contact : userPhoneBook.keySet()) {
                if (contact.equals(contactNameAccess)) {
                    System.out.println("Name: " + contact + "; Phone Number: " + userPhoneBook.get(contact));
                }
            }
                
            phoneBook();
        } else if (phoneBookSelection == 3) {
            System.out.println("Enter the name of the phone contact you wish to remove:");
            String contactNameRemove = userInput2.nextLine();
            userPhoneBook.remove(contactNameRemove);
            System.out.println("Removed.");
            phoneBook();
        } else if (phoneBookSelection == 4) {
            System.out.println("Output:");
            for (String contact : userPhoneBook.keySet()) {
                System.out.println("Name: " + contact + "; Phone Number: " + userPhoneBook.get(contact));
            }
            phoneBook();
        } else if (phoneBookSelection == 0) {
            System.out.println("");
            startup();
        }
    }

    public void popQuiz() {
        System.out.println("");
        System.out.println("Pop Quiz");
        System.out.println("--------");
        int score = 0;
        ArrayList<String> wrongAnswers=  new ArrayList<String>();
        System.out.println("Question #1: Who is the 40th president of the United States?");
        String a1 = userInput2.nextLine().toLowerCase();
        if (a1.equals("ronald reagan")) {
            System.out.println("Correct!");
            score++;
        } else {
            System.out.println("Wrong!");
            wrongAnswers.add("Q1");
        }
        System.out.println("Question #2: What is 7!?");
        int a2 = userInput.nextInt();
        if (a2 == 5040) {
            System.out.println("Correct!");
            score++;
        } else {
            System.out.println("Wrong!");
            wrongAnswers.add("Q2");
        }
        System.out.println("Question #3: What is the period element (full name) corresponding to atomic number 83?");
        String a3 = userInput2.nextLine().toLowerCase();
        if (a3.equals("bismuth")) {
            System.out.println("Correct!");
            score++;
        } else {
            System.out.println("Wrong!");
            wrongAnswers.add("Q3");
        }
        System.out.println("Question #4: What field of physics deals with motion and forces of physical objects?");
        String a4 = userInput2.nextLine().toLowerCase();
        if (a4.equals("mechanics")) {
            System.out.println("Correct!");
            score++;
        } else {
            System.out.println("Wrong!");
            wrongAnswers.add("Q4");
        }
        System.out.println("Question #5: What field of science deals with computation, formation, and automation?");
        String a5 = userInput2.nextLine().toLowerCase();
        if (a5.equals("computer science")) {
            System.out.println("Correct!");
            score++;
        } else {
            System.out.println("Wrong!");
            wrongAnswers.add("Q5");
        }

        System.out.println("");
        System.out.println("Game Complete!");
        System.out.println("Your final score: " + score);
        System.out.println("Questions you got wrong: ");
        if (wrongAnswers.size() == 0) {
            System.out.println("None");
        } else {
            for (String answer : wrongAnswers) {
                System.out.println(answer);
            }
        }

        System.out.println("Play Again? (y/n)");
        String playOrNot = userInput2.nextLine().toLowerCase();
        if (playOrNot.equals("y")) {
            popQuiz();
        } else if (playOrNot.equals("n")) {
            startup();
        }
    }

    public void reverser() {
        System.out.println("");
        System.out.println("Word and Sentence Reversers");
        System.out.println("---------------------------");
        System.out.println("1) Word Reverser");
        System.out.println("2) Sentence Reverser");
        System.out.println("0) Return to main menu");
        int reverserSelection = userInput.nextInt();
        if (reverserSelection == 1) {
            System.out.println("Enter a word:");
            String wordInput = userInput2.nextLine();
            wordReverser(wordInput);
        } else if (reverserSelection == 2) {
            System.out.println("Enter a sentence:");
            String sentenceInput = userInput2.nextLine();
            sentenceReverser(sentenceInput);
        } else if (reverserSelection == 0) {
            startup();
        }
    }

    static void wordReverser(String word) {
        ArrayList<Character> charList = new ArrayList<Character>();
        for (int i = 0; i < word.length(); i++) {
            charList.add(word.charAt(i));
        }
        for (int i = 0; i < (charList.size())/2; i++) {
            char temp = charList.get(i);
            charList.set(i, charList.get(charList.size() - (i + 1)));
            charList.set(charList.size() - (i + 1), temp);
        }
        for (int i = 0; i < charList.size(); i++) {
            System.out.print(charList.get(i));
        }
    }

    static void sentenceReverser(String sentence) {
        for (int i = 0; i < sentence.length(); i++) {
            if (sentence.charAt(i) == ' ') {
                for (int a = 0; a < i; a++) {
                    System.out.print(sentence.charAt(a));
                }
            }
        }
    }

    public void startup() {
        System.out.println("Java Starter Menu");
        System.out.println("-----------------");
        System.out.println("1) Phone Book");
        System.out.println("2) Pop Quiz");
        System.out.println("3) Word and Sentence Reversers");
        System.out.println("4) Animation");
        System.out.println("0) Exit");
        System.out.println("Enter a number 0-4 to select a feature:");
        int selection = userInput.nextInt();
        switch(selection) {
            case 1:
                phoneBook();
                break;
            case 2:
                popQuiz();
                break;
            case 3:
                reverser();
                break;
            case 4:
                //boatAnimation();
                break;
            case 0:
                System.out.println("Leaving...");
                break;
        }
    }
}

class Play {
    public static void main(String[] args) {
        Menu myObj = new Menu();
        myObj.startup();
    }
}

Play.main(null);
```
